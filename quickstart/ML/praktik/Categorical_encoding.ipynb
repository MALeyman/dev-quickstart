```python
def fillna(df):
  # так лучше не делать, лучше с пропусками каждого признака работать отдельно
  for c in df.columns:
    df[c] = df[c].fillna(df[c].mode()[0]) # заменяет все пропуски в колонке на самое частое значение - моду
    # мода может быть не одна, поэтому просто выбрать первую
  return df
```

Категориальный признак — признак, который имеет две или более категорий (значений).

Типы категорий:

**Номинальный**

Не имеет внутреннего порядка в своих категориях. Пол : мужской, женский

**Порядковый**

Имеет порядок внутри категорий. Например, степень образования: бакалавр -> магистр -> доктор


# Подходы к обработке

## One hot encoding

Для **номинальных** категориальных признаков.

**Каждое** значение категории преобразуется в **новый столбец** и ему присваивается значение 1 или 0.

Плюс:

· Простота использования и хорошо подходит для данных с **небольшим** количеством категорий.

Минус:

· Увеличит пространство признаков

```python
df = read_data()
df = fillna(df)
```

```python
# простой встроенный метод пандаса, но
df = pd.get_dummies(df, prefix = ['S'], columns = ['Study Results'], drop_first = True) # drop_first = True
# для бинарных признаков получается 1 колонка лишняя
# df.drop(df.columns[-1], axis=1, inplace=True) # удалит последнюю колонку, и теперь бинарный признак в 1 столбец
df[df.columns[-1:]] = df[df.columns[-1:]].astype(int)
df.head()
```

```python
# свернуть категории обратно можно, если количество столбцов не редактировалось
# иначе обратится некорректно, то есть для бинарных категорий лучше не использовать
df.iloc[:,-1:].idxmax(axis=1)
```

## Ordinal encoding

Для порядковых категориальных признаков.

В этом методе каждому уникальному значению категории присваивается целочисленное значение. красный - 1,зеленый - 2, синий - 3

Числовые категории, не находятся между 0 и 1. Из-за этого Label Encoding не подходит для линейных моделей, где величина признаков играет важную роль. Алгоритмы на деревьях решений от порядка не зависят

Плюсы:

· Простота и понятность реализации

· Широко используется при кодировании данных опросов и исследований.

Минус:

· Нет стандартизированной шкалы интервалов.


Целевое порядковое кодирование

Преобразование признака путем сравнения его с целевой или выходной переменной.

Шаги:

1) Выбрать категориальную переменную.

2) Расчитать среднее целевой переменной и примените его к категориальной переменной.

3) Присвойте более высокие целые значения или более высокий ранг категории с самым высоким средним.

Преимущества:

· Связь между переменной и целью.

· Помогает в более быстром обучении

Недостатки:

· Из-за тесной связи с целевой переменной это часто приводит к переобучению.

Подробно про Target encoding
https://habr.com/ru/articles/545380/

```python
ordinal_label={k:i for i,k in enumerate(ordinal_index, 0)}
ordinal_label
```

```python
df['Status_ordinal_label']=df['Status'].map(ordinal_label)
df.head()
```

## Count / Frequency Encoding



Кодирование количества или частоты

Как следует из названия, в этом методе мы заменим категории на количество наблюдений, которые показывают эту категорию в наборе данных.

Если Индия появляется 56 раз в столбце страны, а Америка появляется 49 раз, мы заменим Индию на 56, а Америку на 49 в столбце страны.

Плюсы:

· Простота реализации

· Нет пространства признаков.

· Хорошо работает с алгоритмами на основе дерева.

Недостатки:

· Не будет обеспечиваться одинаковый вес, если частоты одинаковы.

```python
country_map = df['Study Type'].value_counts().to_dict()
df['Study Type'] = df['Study Type'].map(country_map)
df.head()
```

## Mean Ordinal encoding


Среднее порядковое кодирование

Визуальный вариант целевого порядкового кодирования. Замена полученным средним значением вместо того, чтобы присваивать ей целочисленные значения.

Плюсы:

· Повышает эффективность модели классификации.

· Быстрое получение информации

Минусы:

· Приводит к переобучению

· Может привести к возможной потере значения, если две категории имеют одинаковое среднее значение

```python
mean_ordinal = df.groupby(['Status'])['Phase OE'].mean().to_dict()
mean_ordinal
```

```python
df['Status_mean_ordinal']=df['Status'].map(mean_ordinal)
df.head()
```

## Probability Ratio Encoding

Кодирование отношения вероятности

Этот метод подходит для задач классификации только тогда, когда целевая переменная является двоичной (либо 1, либо 0, либо True или False).

Замена значения отношением вероятности P(1)/P(0).


Плюсы:

· Не расширяет пространство признаков.

· Собирает информацию из признака, что приводит к более прогнозируемым признакам.

Минусы:

· Не определено, когда P(0) = 0.

· Иногда это приводит к переобучению.

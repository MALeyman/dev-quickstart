```python
# Condition, wait() и notify()
import threading
import time

condition = threading.Condition() # блокировка по условию
result_ready = False  # Флаг состояния, указывающий на готовность результата

def preparing_thread():
    global result_ready
    with condition:
        print("Подготовка результата...")
        time.sleep(7)         # Имитация задержки
        result_ready = True   # Устанавливаем флаг состояния в True
        condition.notify()    # Уведомляем о готовности результата

thread = threading.Thread(target=preparing_thread)
thread.start()

with condition:
    while not result_ready:  # Проверяем флаг состояния
        condition.wait()     # Ожидаем уведомления, если результат не готов
    print("Результат готов.")

thread.join() 
```

```python
from concurrent.futures import ThreadPoolExecutor
import threading
import time

# Функция инициализации для потоков
def thread_initializer(greeting):
    print(f"{greeting} от {threading.current_thread().name}")

# Задача, которую будут выполнять потоки
def perform_task(task_number):
    print(f"Задача {task_number} выполняется потоком {threading.current_thread().name}")
    time.sleep(1)
    print(f"Задача {task_number} завершена потоком {threading.current_thread().name}")

# Создание и использование ThreadPoolExecutor
with ThreadPoolExecutor(
    max_workers=2, 
    thread_name_prefix='DemoThread',
    initializer=thread_initializer, # вызывается при старте каждого потока
    initargs=('Привет',) # аргументы для функции
) as executor:
    # Отправка задач на выполнение в пул потоков
    for i in range(4):
        executor.submit(perform_task, i)

# После выхода из блока with пул потоков будет автоматически закрыт, а все задачи завершены
```

# Одновременная загрузка файлов

Функция as_completed() используется для отслеживания задач (Future) в порядке их завершения, не обязательно в том порядке, в котором они были запущены. Это позволяет оперативно реагировать на результаты, поскольку функция блокируется и возвращает каждый завершенный Future по одному, позволяя обрабатывать результаты или исключения, возникшие во время выполнения задач. 

Это делает as_completed() идеальным инструментом для сценариев, где необходимо оперативно реагировать на завершение задач, обрабатывать исключения и управлять порядком выполнения асинхронных операций.

```python
from concurrent.futures import ThreadPoolExecutor, as_completed
import time


# Функция, имитирующая длительную задачу
def long_running_task(seconds):
    print(f"Задача будет выполняться {seconds} секунд.")
    time.sleep(seconds)
    return f"Задача выполнилась за {seconds} секунд."


# Создание пула потоков
with ThreadPoolExecutor(max_workers=3) as executor:
    # Отправка задачи на выполнение
    future_task = executor.submit(long_running_task, 5)

    # Проверка статуса задачи
    if future_task.running():
        print("Задача выполняется...")

    # Получение результата задачи
    print(future_task.result())

    # Добавление обратного вызова к задаче
    future_task.add_done_callback(lambda x: print("Задача завершена!"))
```

```python
from concurrent.futures import ThreadPoolExecutor
import time

def task():
    time.sleep(2)
    return "Задача завершена"

def callback(future):
    print("Callback:", future.result())

with ThreadPoolExecutor() as executor:
    future = executor.submit(task)
    print(future.cancel())  # Попытка отменить задачу
    
    # Продолжение предыдущего примера
    print(future.cancelled())  # Проверяем, была ли задача отменена
    
    # Предполагая, что задача долгая и не была отменена
    print(future.running())  # Проверяем, выполняется ли задача
    
    # Предполагая, что задача завершена или отменена
    print(future.done())  # Проверяем, завершена ли задача
    
    # Предполагая, что future представляет завершенную или неудачную задачу
    try:
        print(future.exception(timeout=1))  # Получаем исключение с таймаутом
    except Exception as e:
        print(e)
        
    future.add_done_callback(callback)  # Добавляем функцию обратного вызова
    
    future.set_result("Результат работы")
    print(future.result())  # Выведет "Результат работы"
```

```python
import os
from urllib.request import urlopen
from concurrent.futures import ThreadPoolExecutor, as_completed

# Функция загрузки данных с URL
def download_url(url):
    try:
        with urlopen(url, timeout=10) as connection:  # Увеличен таймаут для стабильности
            return (connection.read(), url)
    except Exception as e:
        print(f"Ошибка при загрузке {url}: {e}")  # Улучшенное логирование ошибок
        return (None, url)

def save_file(url, data, path):
    filename = os.path.basename(url)
    outpath = os.path.join(path, filename)
    with open(outpath, 'wb') as file:
        file.write(data)
    return outpath

def download_docs(urls, path):
    os.makedirs(path, exist_ok=True)
    with ThreadPoolExecutor(max_workers=40) as executor:  # Ограничение количества потоков для оптимизации
        futures = [executor.submit(download_url, url) for url in urls]
        for future in as_completed(futures):
            data, url = future.result()
            if data is None:
                continue  # Пропускаем неудачные загрузки
            outpath = save_file(url, data, path)
            print(f'Сохранено {url} в {outpath}')
    


# Список URL-адресов для загрузки
URLS = [f'https://asyncio.ru/multithreading/zadachi/6.1/2/{x}.jpg' for x in range(1, 501)]

# Локальный путь для сохранения файлов
PATH = 'downloaded_images'  # Измените на желаемый путь

# Запускаем загрузку всех документов
download_docs(URLS, PATH)
```

# AsyncIO vs ThreadPoolExecutor

## ThreadPoolExecutor
• Модель выполнения: Многопоточная. Использует 
потоки для выполнения задач конкурентно.

 • Функции: Обычные функции.

 • Ожидание задач: Результаты задач получаются через 
объекты Future, возвращаемые методом submit().

 • Запуск задач: Задачи отправляются на выполнение в 
пул потоков с помощью метода submit().

 • Пауза в выполнении: time.sleep(1) - блокирующая пауза, 
поток останавливается.

 • Применение: Подходит для IO-bound и CPU-bound
 задач, но ограничен Global Interpreter Lock (GIL) в 
Python.

```python
import time
from concurrent.futures import ThreadPoolExecutor


# Функция для выполнения задачи в потоке
def my_task(number):
    print(f"Задача {number} начата")
    time.sleep(1)  # Имитация длительной операции
    print(f"Задача {number} завершена")
    return number


# Основная функция для запуска задач в потоках
def main():
    with ThreadPoolExecutor(max_workers=5) as executor:
        # Создание и запуск задач
        futures = [executor.submit(my_task, i) for i in range(5)]

        # Получение результатов задач
        results = [future.result() for future in futures]

    print(f"Результаты: {results}")


# Запуск основной функции
main()
```

## AsyncIO
 • Модель выполнения: Асинхронная. Использует событийный 
цикл (event loop) для управления выполнением.

 • Функции: Асинхронные функции, корутины(с ключевым 
словом async).

 • Ожидание задач: Использует await для ожидания завершения 
асинхронной операции.

 • Запуск задач: Задачи запускаются с помощью asyncio.gather(), 
который эффективно управляет множеством асинхронных 
задач.

 • Пауза в выполнении: await asyncio.sleep(1) - асинхронная 
пауза, не блокирует цикл событий.

 • Применение: Идеально подходит для IO-bound задач, 
особенно когда требуется масштабируемость без создания 
большого количества потоков.

```python
import time
import asyncio

# Определение корутины
async def work():
    # Выполнение действий..
    time.sleep(7)

# Создание и запуск корутины, ожидание её завершения
await work()
```

```python
import asyncio
import time

# Функция для асинхронного выполнения задачи
async def my_task(number):
    print(f"Задача {number} начата")
    await asyncio.sleep(7)  # Имитация длительной операции
    print(f"Задача {number} завершена")
    return number

# Основная функция для запуска асинхронных задач
async def main():
    tasks = [my_task(i) for i in range(5)]  # Создание списка задач
    results = await asyncio.gather(*tasks)  # Запуск задач и ожидание их завершения
    print(f"Результаты: {results}")

# Запуск асинхронного цикла событийa
await main()
```

````python
def bincount_3d(arr, bins=None):
    # Определить максимальное значение
    max_val = arr.max()
    
    # Если не указано количество бинов, вычисляем его автоматически
    if bins is None:
        bins = max_val + 1
    
    # Создать результатный массив
    count = np.zeros((arr.shape[0], arr.shape[1], bins), dtype=np.int64)
    
    # Использовать np.add.at для подсчета
    index3d = np.ones_like(arr) * np.reshape(np.arange(arr.shape[0]), shape=[arr.shape[0], 1, 1])
    index2d = np.ones_like(arr) * np.reshape(np.arange(arr.shape[1]), shape=[1, arr.shape[1], 1])
    np.add.at(count, (index3d, index2d, arr), 1)
    
    return count

# Пример использования
arr_3d = np.random.randint(0, 10, size=(3, 4, 5))
result = bincount_3d(arr_3d)
print(result)
````

## 4. Рассматривая четырехмерный массив, как получить сумму сразу по двум последним осям? (★★★)
`hint: sum(axis=(-2,-1))`

````python
A = np.random.randint(0,10,(3,4,3,4))
# решение путем передачи набора осей (введено в numpy 1.7.0)
sum = A.sum(axis=(-2,-1))
print(sum)
# решение путем преобразования двух последних измерений в одно
# (полезно для функций, которые не принимают кортежи в качестве аргумента axis)
sum = A.reshape(A.shape[:-2] + (-1,)).sum(axis=-1)
print(sum)
````

## 6. Как получить диагональ скалярного произведения? (★★★)
`hint: np.diag`

````python
# Author: Mathieu Blondel

A = np.random.uniform(0,1,(5,5))
B = np.random.uniform(0,1,(5,5))

# Slow 
np.diag(np.dot(A, B)) # Этот метод вычисляет произведение матриц A и B, затем берет диагональ элементов результата.

# Fast 
np.sum(A * B.T, axis=1) # Этот метод умножает транспортированную матрицу B на A, затем суммирует по первой оси.

# Faster 
np.einsum("ij,ji->i", A, B) # Использует einsum для прямого вычисления диагональных элементов произведения матриц.
````

## 11. Как вычислить средние значения, используя скользящее окно по массиву? (★★★)
`hint: np.cumsum, from numpy.lib.stride_tricks import sliding_window_view (np>=1.20.0)`

**«Скользящее окно» (Sliding Window)** — это техника оптимизации, которая используется для решения задач, связанных с обработкой подмассивов или подстрок фиксированной длины в массиве или строке.

````python
# Author: Jaime Fernández del Río

def moving_average(a, n=3) :
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n
Z = np.arange(20)
print(moving_average(Z, n=3))
````

````python

from numpy.lib.stride_tricks import sliding_window_view
````

````python
# Author: Jeff Luo (@Jeff1999)
# NumPy >= 1.20.0


Z = np.arange(20)
print(sliding_window_view(Z, window_shape=3).mean(axis=-1))
````

## 14. Вычислить ранг матрицы (★★★)
`hint: np.linalg.svd, np.linalg.matrix_rank`

````python
# Author: Stefan van der Walt

Z = np.random.uniform(0,1,(10,10))
U, S, V = np.linalg.svd(Z) # Singular Value Decomposition
rank = np.sum(S > 1e-10)
print(rank)

# alternative solution:
# Author: Jeff Luo (@Jeff1999)

rank = np.linalg.matrix_rank(Z)
print(rank)
````

## 15. Рассмотрим набор p матриц с формой (n,n) и набор p векторов с формой (n,1). Как вычислить сумму произведений p матриц сразу? (результат имеет форму (n,1)) (★★★)
`hint: np.tensordot`

````python
# Author: Stefan van der Walt

p, n = 10, 20
M = np.ones((p,n,n))
V = np.ones((p,n,1))
S = np.tensordot(M, V, axes=[[0, 2], [0, 1]])
print(S)

# Это работает, потому что:
# M равно (p,n,nn)
# V равно (p,n,1)
# Таким образом, суммируем по парным осям 0 и 0 (независимо от M и V).,
# и 2 и 1, чтобы остаться с вектором (n,1).
````

## 16. Рассмотрим массив размером 16х16, как получить сумму блоков (размер блока равен 4х4)? (★★★)
`hint: np.add.reduceat, from numpy.lib.stride_tricks import sliding_window_view (np>=1.20.0)`

````python
# Author: Robert Kern

Z = np.ones((16,16))
k = 4
S = np.add.reduceat(np.add.reduceat(Z, np.arange(0, Z.shape[0], k), axis=0),
                                       np.arange(0, Z.shape[1], k), axis=1)
print(S)
````

````python

# alternative solution:
# Author: Sebastian Wallkötter (@FirefoxMetzger)

Z = np.ones((16,16))
k = 4

windows = np.lib.stride_tricks.sliding_window_view(Z, (k, k))
S = windows[::k, ::k, ...].sum(axis=(-2, -1))
S
````

````python
# Author: Jeff Luo (@Jeff1999)

Z = np.ones((16, 16))
k = 4
print(sliding_window_view(Z, window_shape=(k, k))[::k, ::k].sum(axis=(-2, -1)))
````

## 17. Как реализовать игру жизни с использованием массивов numpy? (★★★)

````python
# Author: Nicolas Rougier

def iterate(Z):
    # Посчитать соседей
    N = (Z[0:-2,0:-2] + Z[0:-2,1:-1] + Z[0:-2,2:] +
         Z[1:-1,0:-2]                + Z[1:-1,2:] +
         Z[2:  ,0:-2] + Z[2:  ,1:-1] + Z[2:  ,2:])

    # Apply rules
    birth = (N==3) & (Z[1:-1,1:-1]==0)
    survive = ((N==2) | (N==3)) & (Z[1:-1,1:-1]==1)
    Z[...] = 0
    Z[1:-1,1:-1][birth | survive] = 1
    return Z

Z = np.random.randint(0,2,(50,50))
for i in range(100): Z = iterate(Z)
print(Z)
````
